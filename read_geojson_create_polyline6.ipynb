```python
# # This is using the polyline6 format
# # This one is WORKING!!!!! It worked finally when I used the Polyline6 format instead of Polyline5, then replacing the \ with \\


import geojson
import polyline
import codecs

def convert_coordinates(feature):
    coordinates = feature['geometry']['coordinates'][0]
    converted_coordinates = [(lat, lon) for lon, lat in coordinates]
    # print(converted_coordinates)
    polyline_encoded = polyline.encode(converted_coordinates, 6)
    return polyline_encoded

def write_to_text_file(uid, polyline6, output_directory):
    filename = f"{output_directory}/{uid}.txt"
    with open(filename, 'w', encoding='utf-8') as text_file:
        # Replace all '\' with '\\'
        polyline6_escaped = polyline6.replace('\\', '\\\\')
        text_file.write(polyline6_escaped)

def process_geojson(input_file, output_directory):
    with open(input_file, 'r', encoding='utf-8') as file:
        geojson_data = geojson.load(file)

    for feature in geojson_data['features']:
        uid = feature['properties']['uid']
        polyline6 = convert_coordinates(feature)

        write_to_text_file(uid, polyline6, output_directory)

if __name__ == "__main__":
    input_file = "Driving_Route_line.geojson"
    output_directory = "D:\Code\Python\MSIL_Route_Analysis\Driving_Route_line"

    process_geojson(input_file, output_directory)
```
